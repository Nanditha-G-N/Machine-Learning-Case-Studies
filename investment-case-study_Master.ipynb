# Task: Import Files and resolve encoding

*__SubTask: Load companies data into companies data frame__*

```python
# code for importing the companies.csv file. ipynb file and the source file in the same location 
companies = pd.read_csv("companies.txt",sep='\t',encoding="ISO-8859-1",header=(0)) 
companies.head(5)
```

*__SubTask: Load rounds data into rounds2 data frame__*

```python
# code for importing the rounds2.csv file. ipynb file and the source file in the same location 
rounds2 = pd.read_csv("rounds2.csv",encoding="ISO-8859-1",header=(0)) 
rounds2.head(5)
```

*__SubTask: Resolve encoding issues__*

*__Outcome: updated rounds2 and companies file__*

```python
# code for resolving encoding issues in company_permalink column in rounds2 dataframe
rounds2['company_permalink'] = rounds2.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
rounds2.head(5)
```

```python
# code for resolving encoding issues in permalink and name column in companies dataframe
companies['permalink'] = companies.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
companies['name'] = companies.name.str.encode('utf-8').str.decode('ascii', 'ignore')
companies.head(5)
```

```python
# code for testing if the encoding has been resolved
companies[companies['homepage_url']=='http://inovatiff.com']
```

*__Comments: since the file is only a reference file, the below code is only to undersatnd the data. Also the country name doesnt have country code associated, hence this will not be used to merge with rounds2 data__*

```python
# reading the pdf file
pdf_object = open('Countries_where_English_is_an_official_language.pdf', 'rb')
pdfobject = pdf.PdfFileReader(pdf_object)
countries = pdfobject.getPage(0).extractText()
countries.split('\n')
```

# Task 18: Map each company's primary sector to one of the 8 main sectors defined in the mappings.csv
### Outcome: Merged data frame that maps the primary sector to the main sector

```python
master_frame_venture = master_frame_venture[master_frame_venture['country_code'].isin(['USA','GBR','IND'])]
master_frame_venture.country_code.unique()
```

```python
master_frame_venture["top_category"] = master_frame_venture['category_list'].str.split("|", n = 1, expand = True)[0]
master_frame_venture.top_category.unique()
```

```python
df_mapping = pd.read_csv("mapping.csv")
df_category_err = df_mapping[(df_mapping['category_list'].str.contains("0",na=False))&(~df_mapping['category_list'].str.contains("2.0",na=False))]
df_mapping.loc[(df_mapping['category_list'].str.contains("0",na=False))&(~df_mapping['category_list'].str.contains("2.0",na=False)),['category_list']] = df_category_err['category_list'].str.replace("0","na")
```

```python
df_mapping[df_mapping['category_list'].str.contains("2.0",na=False)]
```

```python
column_value = np.array(df_mapping.columns.tolist())
column_value = column_value.reshape(1,10)
df_mapping_list = pd.melt(df_mapping, id_vars=['category_list'], value_vars=print(column_value))
df_mapping_list = df_mapping_list[df_mapping_list["value"]==1]
df_mapping_list.head(10)
```

```python
df_mapping_list["sector"] = df_mapping_list["variable"] 
df_mapping_list = df_mapping_list.drop(["variable","value"],axis=1)
df_mapping_list[df_mapping_list["category_list"]=='natural gas Uses']
```
