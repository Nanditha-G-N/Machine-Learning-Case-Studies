```python
# Import the file
df_mapping = pd.read_csv("mapping.csv")
df_mapping
```

*__Review and Clean Data__*
> - Data Issue: Few rows have 0 instead of "na" for example "A0lytics" instead of "Analytics" <br>
> - Need to replace 0 where applicable with "na" <br>
> - Ensure correct 0s are not replaced


```python
# replace 0 with "na", ensure Enterprise 2.0 is kept intact
df_category_err = df_mapping[(df_mapping['category_list'].str.contains("0",na=False))&(~df_mapping['category_list'].str.contains("2.0",na=False))]
df_mapping.loc[(df_mapping['category_list'].str.contains("0",na=False))&(~df_mapping['category_list'].str.contains("2.0",na=False)),['category_list']] = df_category_err['category_list'].str.replace("0","na")
```

```python
# Test for corrected values
df_mapping[(df_mapping['category_list'].str.contains("2.0",na=False)) | (df_mapping['category_list']=="Analytics")]
```

*__Subtask: Convert the cross tabular data to identify the mapping where value is 1__*

```python
# Convert the crosstab to list and filter the data for value =1 to identify the mapping
column_value = np.array(df_mapping.columns.tolist())
column_value = column_value.reshape(1,10)
df_mapping_list = pd.melt(df_mapping, id_vars=['category_list'], value_vars=print(column_value))
df_mapping_list = df_mapping_list[df_mapping_list["value"]==1]
df_mapping_list.head(10)
```

```python
# Rename the columns as per the assignment and drop unnecessary columns
df_mapping_list["main_sector"] = df_mapping_list["variable"] 
df_mapping_list = df_mapping_list.drop(["variable","value"],axis=1)
```
